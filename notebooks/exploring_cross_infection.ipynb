#### Scenarios

**Scenario 1:** High transmission rates and low recovery rates for both variants

If plotting cross-infection reports still infected at the end of the run, the variants haven't burned out yet and t_total could be increased to fully capture the epidemic

```python
#where variant 2 emerges and takes over
params = [0.78, 0.85, 0.22, 0.15] #beta1, beta2, gamma1, gamma2
y0 = [940, 60, 0, 0, 0, 0, 0, 0, 0] #
t, y = ci.simulate_cross_infection(params, y0, t_emerge=100)

ci.plot_cross_infection(t, y)
```

Here, I'm working with a population of 1000. Both variants have high transmission rates and lower recovery rates. Almost everyone gets variant 1, then 2.

**Scenario 2:** Lowering variant 1's transmission rate and raising variant 2's recovery rate.

```python
#where variant 2 emerges and takes over
params = [0.45, 0.85, 0.22, 0.55] #beta1, beta2, gamma1, gamma2
y0 = [940, 60, 0, 0, 0, 0, 0, 0, 0] #
t, y = ci.simulate_cross_infection(params, y0, t_emerge=100)

ci.plot_cross_infection(t, y)
```

Lowering the transmission rate of the first variant and raising the recovery rate of the second variant. Lowering the first variant's transmission rate meant that the green line did not peak nearly as high or as fast as the previous model. The susceptible population held pretty steady until the second variant emerged.
