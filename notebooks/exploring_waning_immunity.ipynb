```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../models')
#imports
import numpy as np
import matplotlib.pyplot as plt
from waning_immunity import sim, simulator, plot_simulation
```

**Exploring different scenarios**

Scenario 1 - No waning

```python
t = np.linspace(0, 500, num=1000) #1000 time 

beta = 0.3
gamma = 0.2
omega = 0.00 #waning rate

params = [beta, gamma, omega]

# initial conditions vector
y0 = [990, 10, 0] # 500 susceptible, 70 infected, 0 recovered people

y = simulator(params, y0, t) #simulate the SIR model with waning immunity
plot_simulation(t, y, params) #plot the simulation results
```

Turning off waning reduces the model to a normal SIR model; it burns out and leaves infections at 0.

Scenario 2 - Waning immunity through damped oscillations

```python
beta = 0.3
gamma = 0.2
omega = 0.02 #waning rate

params = [beta, gamma, omega]

# initial conditions vector
y0 = [990, 10, 0] # 500 susceptible, 70 infected, 0 recovered people

y = simulator(params, y0, t) #simulate the SIR model with waning immunity

plot_simulation(t, y, params) #plot the simulation results
```

Scenario 3 - Slowing down waning rate

```python
beta = 0.3
gamma = 0.2
omega = 0.009 #waning rate

params = [beta, gamma, omega]

# initial conditions vector
y0 = [990, 10, 0] # 990 susceptible, 10 infected, 0 recovered people

y = simulator(params, y0, t) #simulate the SIR model with waning immunity
plot_simulation(t, y, params) #plot the simulation results
```

The slower waning rate keeps people in the R class longer, so fewer are actively infected at a steady rate. This new value of 0.009 widens the wave period and lowers the endemic infection level from ~30 to ~14,  which is the steady number of infected people.
