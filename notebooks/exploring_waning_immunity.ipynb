## Exploring Waning Immunity
**Author:** Karen Guzman

**Description:** Exploring different values of susceptible, infected, and recovered people with varying rates beta, gamma, and omega. The waning immunity parameter, omega, is the rate at which recovered individuals lose immunity and become susceptible again.

**Date:** 6/18/2026

**Credit:** The basic SIR model was based off of the following Youtube video https://www.youtube.com/watch?v=cbXCyO_F2v8. Waning immunity was then added to this starter code.

```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../models')
#imports
import numpy as np
import matplotlib.pyplot as plt
from waning_immunity import sim, simulator, plot_simulation
```

```python
# define space of time points we'll be solving over
t = np.linspace(0, 500, num=1000) #1000 time points/samples

# parameters, here beta is > than gamma to ensure that the disease spreads
beta = 0.04
gamma = 0.01
omega = 0.005 #waning rate

params = [beta, gamma, omega]

# initial conditions vector
y0 = [100, 1, 0] # 100 susceptible, 1 infected, 0 recovered people
```

```python
y = simulator(params, y0, t) #simulate the SIR model with waning immunity

plot_simulation(t, y, beta, gamma) #plot the simulation results
```
